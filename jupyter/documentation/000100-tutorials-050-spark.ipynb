## Spark kurulumu

1. jdk'yi download et
```
wget --no-cookies --no-check-certificate --header "Cookie: gpw_e24=http%3A%2F%2Fwww.oracle.com%2F; oraclelicense=accept-securebackup-cookie" "http://download.oracle.com/otn-pub/java/jdk/8u171-b11/512cd62ec5174c3487ac17c61aaa89e8/jdk-8u171-linux-x64.tar.gz"
```

1. jdk'yi unzip edip .bashrc'ye tanit.
```
```

1. install scala build tool (sbt)
```
$> curl https://bintray.com/sbt/rpm/rpm | sudo tee /etc/yum.repos.d/bintray-sbt-rpm.repo
$> sudo yum install sbt
```

1. install scala per project (it is installed per project, not system wide)
```
$> pip install <paket adi>
```

1. cd to a root folder and run command below to create a scala project
```
$> sbt new scala/hello-world.g8
```

1. run command below to get into scala shell
```
$> cd hello-world; sbt
```

1. run command below in scala shell to run scala code (~run runs all scalas in subfolders)
```
$> ~run
```

1. install apache spark
```
$> wget http://apache.mirror.anlx.net/spark/spark-2.3.0/spark-2.3.0-bin-hadoop2.7.tgz
$> tar xvf spark-2.3.0-bin-hadoop2.7.tgz
$> echo export PATH=<PWD>/spark-2.3.0/bin:$PATH >> $HOME/.bashrc
$> . $HOME/.bashrc
```
1. run spark shell
```
$> spark-shell
```

## Spark shell ve python3.6

## Counting the number of messages in a kafka's topic

topic name is dashboard. broker is localhost:9092
```
bin/kafka-run-class.sh kafka.tools.GetOffsetShell --broker-list localhost:9092 --topic dashboard --time -1
```

sum all in partitions

```
bin/kafka-run-class.sh kafka.tools.GetOffsetShell --broker-list localhost:9092 --topic dashboard --time -1 | while IFS=: read topic_name partition_id number; do echo "$number"; done | paste -sd+ - |bc
```
