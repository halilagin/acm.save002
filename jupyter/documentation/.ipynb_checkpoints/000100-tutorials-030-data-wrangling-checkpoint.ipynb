#  Veri temizleme ile ilgili bilgiler


## Veri temizleme kod ornegi

ACM projesinde acm/datasource ve acm/venvs/Alo185DataAnalysis dizinlerinde ilgili kod ornekleri mevcuttur.

Veri temizleme icin asagidaki adimlari takip edecegiz.


1. Python kod standardina uygun datasource sinifi yaratilir (acm/datasource altinda)
1. acm/venvs altinda calisma ortami yaratilir
1. configrasyon seviyesi acm/venvs altindaki python dosyasi ile yapilir
1. acm paketi ve venvs paketleri soyle ayrilir:
    1. acm paketleri aPI'yi tanimlar
    1. venv python dosyalari tanimlanmis API'nin kullanimini yapan kodu yazar.
    1. veri temizleme islemlerinde data source disinda veri yaratip pandas ve numpy arraylary yaratma islemi acm paketlerinde yapilir (kod ornegi acm/datasource altinda var)
    1. Veri temizleme islemleri venv altinda yapilir
1. Veriler HTML decode edilir ve UTF-8 formatinda saklanir
1. Veri temizleme aasamasinda ornegin utf-8 donusumu ardindan html decode yapilacaksa, ara ciktilar dosyalarda saklanir. utf-8 donusumunun ciktisi bir dosyada saklanir, bir sonraki donusturulmus data ayri bir dosyada saklanir.
1. Tum veri kaynaklarinin orjinalleri saklanir ve bir kopya uzerinden calisilir
1. Tum veri kaynaklarinin isletim sistemi yollari yaml dosyalarinda saklanir.  acm/venvs/Alo185DataAnalysis projesine bakilabilir.
1. Veri temizleme islemleri once kucuk dosya parcalarinda (100 satir ornegin), basari elde edilmisse ardindan orjinal dosyada yapilir

        def unescapeHtmlEncode(self, finPath, foutPath):
                text_=None
                with io.open(finPath, 'r', encoding="utf-8") as fin:
                    text_ = fin.read()

                text_ = html.unescape(text_)
                with io.open(foutPath, 'w', encoding="utf-8") as fout:
                    fout.write(text_)

        def extractHtmlText(self, finPath, foutPath):#use beatifulsoup to gettext in <p><p> tag
                pass
                with open(foutPath, "w", encoding="utf-8") as fout:
                        writer = csv.DictWriter(fout, fieldnames=self.csvFields,delimiter=';')
                        writer.writeheader()
                        with open(finPath,"r",  encoding='utf-8') as fin:
                                reader = csv.DictReader(fin,delimiter=';')
                                fields = reader.fieldnames
                                print(fields)
                                try: 
                                        i=0
                                        for row in reader:
                                                i=i+1
                                                if i==11:
                                                        break
                                                #soup = BeautifulSoup(row["f10"], 'html.parser')
                                                #row["f10"]=soup.get_text() #text extracted
                                                writer.writerow(row)
                                except:
                                        print("print error")



        def cleanDataSource(self):

                #phase010
                phase010_finPath = self.config.acm.datasource.alo.path10rows
                phase010_foutPath = self.config.acm.datasource.alo.path_phase010_htmldecoded
                self.unescapeHtmlEncode(phase010_finPath, phase010_foutPath)

                #phase020
                phase020_finPath = phase010_foutPath
                phase020_foutPath = self.config.acm.datasource.alo.path_phase020_htmltextextracted
                self.extractHtmlText(phase020_finPath,phase020_foutPath)

        def start(self):
                print("started")
                self.cleanDataSource()
                



c_ = VenvAlo185DataAnalysis()
